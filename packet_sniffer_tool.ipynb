```python
from scapy.all import sniff, Ether, IP, TCP, UDP

def packet_handler(packet):
    if Ether in packet:
        eth_layer = packet[Ether]
        src_mac = eth_layer.src
        dst_mac = eth_layer.dst

        if IP in packet:
            ip_layer = packet[IP]
            src_ip = ip_layer.src
            dst_ip = ip_layer.dst

            if TCP in packet:
                protocol = "TCP"
            elif UDP in packet:
                protocol = "UDP"
            else:
                protocol = "Other"

            print(f"Source MAC: {src_mac} | Destination MAC: {dst_mac}")
            print(f"Source IP: {src_ip} | Destination IP: {dst_ip} | Protocol: {protocol}")
            print("-" * 40)

# Sniff packets on interface 'eth0' (change as needed)
sniff(iface="eth0", prn=packet_handler, count=10)
```
